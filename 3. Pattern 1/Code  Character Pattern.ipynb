# Code : Character Pattern
Code : Character Pattern<br>
<br>
Print the following pattern for the given N number of rows.<br>
<br>
Pattern for N = 4<br>
A<br>
BC<br>
CDE<br>
DEFG<br>
<br>
Input format :<br>
Integer N (Total no. of rows)<br>
<br>
Output format :<br>
Pattern in N lines<br>
<br>
Constraints<br>
0 <= N <= 13<br>
<br>
Sample Input 1:<br>
5<br>
Sample Output 1:<br>
A<br>
BC<br>
CDE<br>
DEFG<br>
EFGHI<br>
Sample Input 2:<br>
6<br>
Sample Output 2:<br>
A<br>
BC<br>
CDE<br>
DEFG<br>
EFGHI<br>
FGHIJK

```python
n=int(input())
currRow=1
while currRow<=n:
    currCol=1
    char=ord('A')+currRow-1
    while currCol<=currRow:
        print(chr(char+currCol-1),end="")
        currCol+=1
    print()
    currRow+=1
```
