# Code : Triangular Star Pattern
Print the following pattern for the given N number of rows.<br>
Pattern for N = 4<br>
![](https://imgur.com/download/SrTsgrS)<br>
<br>
Note : There are no spaces between the stars.<br>
Input format :<br>
Integer N (Total no. of rows)<br>
Output format :<br>
Pattern in N lines<br>
Constraints<br>
0 <= N <= 50<br>
Sample Input 1:<br>
5<br>
Sample Output 1:<br>
![](https://imgur.com/download/B1x6IkM)<br>
Sample Input 2:<br>
6<br>
Sample Output 2:<br>
![](https://imgur.com/download/iwTrMTd)

```python
n=int(input())
i=1
while i<=n:
    star=1
    while star<=i:
        print("*",end="")
        star+=1

    print("")
    i=i+1
```
