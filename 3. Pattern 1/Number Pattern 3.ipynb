# Number Pattern 3
Print the following pattern for the given N number of rows.<br>
<br>
Pattern for N = 4<br>
1<br>
11<br>
121<br>
1221<br>
<br>
Input format :<br>
Integer N (Total no. of rows)<br>
<br>
Output format :<br>
Pattern in N lines<br>
<br>
Sample Input :<br>
5<br>
Sample Output :<br>
1<br>
11<br>
121<br>
1221<br>
12221

```python
n=int(input())
i=1
print(1)
while i<=n-1:
    print("1",end="")
    num_2=2
    while num_2<=i: #no of Colums
        print("2",end="")
        num_2+=1
    print("1")
    i=i+1
```
