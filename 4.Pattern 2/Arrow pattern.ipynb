# Arrow pattern
Print the following pattern for the given number of rows.
<br>Assume N is always odd.
<br>Note : There is space after every star.
<br><b>Pattern for N = 7</b><br>

[![Arrow pattern](https://shareimage.xyz/images/2021/10/11/NMsMCn7---Imgur-4.png)](https://shareimage.xyz/image/iegGX)<br>
Note-Dote(.) represent the spaces <br>
Input format :<br>
Integer N (Total no. of rows)<br>
<br>
Output format :<br>
Pattern in N lines<br>
<br>
Sample Input :<br>
11<br>
Sample Output :<br>
[![](https://shareimage.xyz/images/2021/10/11/undefined---Imgur.md.png)](https://shareimage.xyz/image/iehrw)

```python
n = int(input())
i = 1
n1 = (n - 1) / 2
n = (n + 1) / 2
while i <= n:
    spaces = 2
    while spaces <= i:
        print(" ", end="")
        spaces += 1
    star = 1
    while star <= i:
        print("* ", end="")
        star += 1
    print("")
    i = i + 1

i = 1
g = 0
while i <= n1:
    num_1 = n1 - 1
    while num_1 >= i:
        print(" ", end="")
        num_1 -= 1

    num_1 = n1 - 1
    while num_1 >= g:
        print("* ", end="")
        num_1 -= 1
    print()
    i = i + 1
    g = g + 1
```
