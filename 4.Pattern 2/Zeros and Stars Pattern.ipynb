# Zeros and Stars Pattern
Print the following pattern<br>
Pattern for N = 4<br>
![ZerosandStarsPattern](https://imgur.com/download/iUdwaMY)<br>
Input Format : N (Total no. of rows)<br>
<br>
Output Format : Pattern in N lines<br>
<br>
Sample Input 1 :<br>
3<br>
Sample Output 1 :<br>
![](https://imgur.com/download/VZgJBG8)<br>
Sample Input 2 :<br>
5<br>
Sample Output 2 :<br>
![](https://imgur.com/download/zEhIMUr)

```python
n=int(input())
start=1
end=2*n+1
mid=n+1
row=1
while(row<=n):
    column=1
    while column<=(2*n+1):
        if column==start or column==end or column==mid:
            print("*",end='')
        else:
            print("0",end="")
        column=column+1
    start=start+1
    end=end-1
    row=row+1
    print()
```
