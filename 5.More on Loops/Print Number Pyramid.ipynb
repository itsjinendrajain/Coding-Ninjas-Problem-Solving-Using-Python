# Print Number Pyramid
Print the following pattern for a given n.<br>
<br>
**Note - Dot Represent Spaces**<br>
For eg. N = 6<br>
![](https://imgur.com/download/p03f2fS)
<br>
Sample Input 1 :<br>
4<br>
Sample Output 1 :<br>
![](https://imgur.com/download/VSpXQ0L)

```python
n=int(input())
for i in range(1,n+1):
    for j in range(1,i):
        print(" ",end="")
    for l in range(i,n+1):
        print(l,end="")
    print()
for a in range(1,n):
    for s in range(n-1-a,0,-1):
        print(" ",end="")
    for k in range(n-a,n+1):
        print(k,end="")
    print()
```
