# Code Merge Two Sorted Arrays
You have been given two sorted arrays/lists(ARR1 and ARR2) of size N and M respectively, merge them into a third array/list such that the third array is also sorted.<br>
<br>
Input Format :<br>
The first line contains an Integer 't' which denotes the number of test cases or queries to be run. Then the test cases follow.<br>
<br>
First line of each test case or query contains an integer 'N' representing the size of the first array/list.<br>
<br>
Second line contains 'N' single space separated integers representing the elements of the first array/list.<br>
<br>
Third line contains an integer 'M' representing the size of the second array/list.<br>
<br>
Fourth line contains 'M' single space separated integers representing the elements of the second array/list.<br>
<br>
Output Format :<br>
For each test case, print the sorted array/list(of size N + M) in a single row, separated by a single space.<br>
<br>
Output for every test case will be printed in a separate line.<br>
<br>
Constraints :<br>
1 <= t <= 10^2<br>
0 <= N <= 10^5<br>
0 <= M <= 10^5<br>
Time Limit: 1 sec <br>
Sample Input 1 :<br>
1<br>
5<br>
1 3 4 7 11<br>
4<br>
2 4 6 13<br>
Sample Output 1 :<br>
1 2 3 4 4 6 7 11 13 <br>
Sample Input 2 :<br>
2<br>
3<br>
10 100 500<br>
7<br>
4 7 9 25 30 300 450<br>
4<br>
7 45 89 90<br>
0<br>
Sample Output 2 :<br>
4 7 9 10 25 30 100 300 450 500<br>
7 45 89 90

```python
def mergesorted(arr1, arr2):
    i = 0
    j = 0
    len1 = len(arr1)
    len2 = len(arr2)
    arr = []
    while ((i < len1) and (j < len2)):
        if (arr1[i] < arr2[j]):
            arr.append(arr1[i])
            i += 1
        else:
            arr.append(arr2[j])
            j += 1
    while (i < len1):
        arr.append(arr1[i])
        i += 1
    while (j < len2):
        arr.append(arr2[j])
        j += 1
    return arr


loop = int(input())
for l in range(loop):
    arr1 = []
    arr2 = []
    ele1=int(input())
    if ele1!=0:
        arr1=[int(c) for c in input().split()]
    ele2=int(input())
    if ele2!=0:
        arr2=[int(v) for v in input().split()]
    arr=mergesorted(arr1,arr2)
    print(*arr)
```
