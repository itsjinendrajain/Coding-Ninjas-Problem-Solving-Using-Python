# Push Zeros to end
You have been given a random integer array/list(ARR) of size N. You have been required to push all the zeros that are present in the array/list to the end of it. Also, make sure to maintain the relative order of the non-zero elements.<br>
<br>
Note:<br>
Change in the input array/list itself. You don't need to return or print the elements.<br>
<br>
You need to do this in one scan of array only. Don't use extra space.<br>
<br>
<br>
Input format :<br>
The first line contains an Integer 't' which denotes the number of test cases or queries to be run. Then the test cases follow.<br>
<br>
First line of each test case or query contains an integer 'N' representing the size of the array/list.<br>
<br>
Second line contains 'N' single space separated integers representing the elements in the array/list.<br>
<br>
Output Format :<br>
For each test case, print the elements of the array/list in the desired order separated by a single space.<br>
<br>
Output for every test case will be printed in a separate line.<br>
<br>
Constraints :<br>
1 <= t <= 10^2<br>
0 <= N <= 10^5<br>
Time Limit: 1 sec<br>
<br>
Sample Input 1:<br>
1<br>
7<br>
2 0 0 1 3 0 0<br>
Sample Output 1:<br>
2 1 3 0 0 0 0<br>
 Explanation for the Sample Input 1 :<br>
All the zeros have been pushed towards the end of the array/list. Another important fact is that the order of the non-zero elements have been maintained as they appear in the input array/list.<br>
Sample Input 2:<br>
2<br>
5<br>
0 3 0 2 0<br>
4<br>
9 0 0 8 2<br>
Sample Output 2:<br>
3 2 0 0 0<br>
9 8 2 0 0

```python
loop=int(input())
for i in range(loop):
    no_element=input()
    n=input().split()
    n.sort(key='0'.__eq__)
    for k in range(len(n)):
        print(n[k],end=' ')
    print()
```
