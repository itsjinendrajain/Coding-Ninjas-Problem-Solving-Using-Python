# Sum of Two Arrays
Two random integer arrays/lists have been given as ARR1 and ARR2 of size N and M respectively. Both the arrays/lists contain numbers from 0 to 9(i.e. single digit integer is present at every index). The idea here is to represent each array/list as an integer in itself of digits N and M.<br>
You need to find the sum of both the input arrays/list treating them as two integers and put the result in another array/list i.e. output array/list will also contain only single digit at every index.<br>
Note:The sizes N and M can be different. <br>
<br>
Output array/list(of all 0s) has been provided as a function argument. Its size will always be one more than the size of the bigger array/list. Place 0 at the 0th index if there is no carry. <br>
<br>
No need to print the elements of the output array/list.<br>
<br>
Input format :<br>
The first line contains an Integer 't' which denotes the number of test cases or queries to be run. Then the test cases follow.<br>
<br>
First line of each test case or query contains an integer 'N' representing the size of the first array/list.<br>
<br>
Second line contains 'N' single space separated integers representing the elements of the first array/list.<br>
<br>
Third line contains an integer 'M' representing the size of the second array/list.<br>
<br>
Fourth line contains 'M' single space separated integers representing the elements of the second array/list.<br>
<br>
Output Format :<br>
For each test case, print the required sum of the arrays/list in a row, separated by a single space.<br>
<br>
Output for every test case will be printed in a separate line.<br>
<br>
Constraints :<br>
1 <= t <= 10^2<br>
0 <= N <= 10^5<br>
0 <= M <= 10^5<br>
Time Limit: 1 sec <br>
Sample Input 1:<br>
1<br>
3<br>
6 2 4<br>
3<br>
7 5 6<br>
Sample Output 1:<br>
1 3 8 0<br>
Sample Input 2:<br>
2<br>
3<br>
8 5 2<br>
2<br>
1 3<br>
4<br>
9 7 6 1<br>
3<br>
4 5 9<br>
Sample Output 2:<br>
0 8 6 5<br>
1 0 2 2 0

```python
def sumArray(arr1,n,arr2,m,output):
    i=n-1
    j=m-1
    carry=0
    l=max(n,m)+1
    while i>=0 and j>=0:
        num=arr1[i]+arr2[j]+carry
        s=num%10
        carry=num//10
        output[l-1]=s
        l=l-1
        i=i-1
        j=j-1
    while i>=0:
        num=arr1[i]+carry
        s=num%10
        carry=num//10
        output[l-1]=s
        l=l-1
        i=i-1
    while j>=0:
        num=arr2[j]+carry
        s=num%10
        carry=num//10
        output[l-1]=s
        l=l-1
        j=j-1
    if carry!=0:
        output[0]=carry
        
    
    
t=int(input())
for i in range(t):
    n=int(input())
    if n==0:
        print(0)
        break
    arr1=[int(x) for x in input().split()]
    m=int(input())
    arr2=[int(x) for x in input().split()]
    size=max(n,m)+1
    output=[0 for i in range(size)]
    sumArray(arr1,n,arr2,m,output)
    print(*output)
    print()
```
